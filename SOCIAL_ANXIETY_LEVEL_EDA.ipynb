### Categorizing Anxiety Level into Low, Medium, and High Classes ##

```python
def classification_anxiety_level(num):
    if num<=3:
        return "Low"
    elif num<=7:
        return "Medium"
    else:
        return "High"
        
```

```python
df["Anxity level classification"]= y.apply(classification_anxiety_level)
```

### Social Anxiety Level Classification Based on Smoking ###

```python
df['Smoking'].value_counts()
```

```python
df["Smoking"]=df["Smoking"].map({
    "Yes":1,
    "No":0
}
    )
```

```python
c=pd.crosstab(df["Smoking"],df["Anxity level classification"])
c.plot(kind='bar', color=["#0077b6","#ade8f4","#00b4d8"]);
plt.title("Anxiety level classification among the Smokers vs. Nonsmokers");
plt.xticks(rotation=0)
```
