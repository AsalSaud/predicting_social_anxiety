```python
y = df["Anxiety Level (1-10)"]
y
```

```python
def classification_anxity_level(num):
    if num<=3:
        return "Low"
    elif num<=7:
        return "Medium"
    else:
        return "High"
        
```

```python
df["Anxity level classification"]= y.apply(classification_anxity_level)
```

# Our Insights #
**The Key factors correlated with anxiety levels**
- Stress level, therapy sessions, and sleep hours show significant correlations with anxiety levels. 
- Stress level and therapy sessions show positive correlations, while sleep hours show a negative correlation. 
- Physical activity and diet quality show weaker negative correlations.
-----
**There are visible outliers in the data** 

- The box plots generated in the analysis visually identify potential outliers in several numerical variables.

-------
## Key Findings ##
* **Strong Correlation with Stress:** Anxiety level exhibits a strong positive correlation with stress level (0.67).
* **Correlation with Therapy:** Anxiety level shows a positive correlation with the number of therapy sessions (0.52).
* **Negative Correlation with Sleep:** Anxiety level demonstrates a negative correlation with sleep hours (-0.49).
* **Potential Outliers:** Box plots reveal potential outliers in several numerical variables
* **No Missing Values:** The dataset has no missing values.
